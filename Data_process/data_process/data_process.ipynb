```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("./benchmark.csv")

# Count the number of unique species
species_counts = df['species'].value_counts()

# Plot the bar chart for the number of different species
plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Different Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=45)

# Add count numbers on top of each bar
for i, count in enumerate(species_counts):
    plt.text(i, count + 0.2, str(count), ha='center', va='bottom')

plt.show()

label_species_counts = df.groupby(['label', 'species']).size().unstack(fill_value=0)

ax = label_species_counts.plot(kind='bar', stacked=False, figsize=(12, 6), width=0.8)
plt.xlabel('Label')
plt.ylabel('Count')
plt.title('Number of Species for Each Label')
plt.legend(title='Species', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)  # labelx

# 
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))


plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
from configs import DATA_PATH, VOCAB_PATH0, VOCAB_PATH1, VOCAB_PATH2, REBUILD_DATA_INDE1_PATH2, REBUILD_DATA_INDE1_PATH1, REBUILD_DATA_INDE1_PATH0, LE_PATH, TOKEN_LEN, PART_NUM, PART_LEN
from utils import save_pkl, save_json, load_json


def label_encoder(y, label_model_path):
    le = LabelEncoder()
    labels = le.fit_transform(y)
    save_pkl(label_model_path, le)
    print("The classes of labels: ", le.classes_)
    return le, labels


class Vocabulary(object):
    def __init__(self):
        self.token2idx = {"<K>": 0}
        self.idx2token = {0: "<K>"}
        self.idx = 1

    def add_token(self, token):
        if token not in self.token2idx:
            self.token2idx[token] = self.idx
            self.idx2token[self.idx] = token
            self.idx += 1

    def add_seq(self, seq):
        for token in seq.split():
            self.add_token(token)

    # transform sequence into number
    def seq2ver(self, seq):
        seq_idxs = [self.token2idx.get(token, self.token2idx["<K>"]) for token in seq.split()]
        return seq_idxs

    def save_dict(self, dict_path):
        save_json(dict_path, self.token2idx)

    def load_dict(self, dict_path):
        self.token2idx = load_json(dict_path)
        self.idx2token = {idx: token for token, idx in self.token2idx.items()}
        self.idx = len(self.token2idx)

    def __call__(self, token):
        return self.token2idx.get(token, self.token2idx["<K>"])

    def __len__(self):
        return self.idx


if __name__ == "__main__":
    data = pd.read_csv("./independent_Li.csv")
    data1 = pd.read_csv("./benchmark_all.csv")

    vocabulary_obj = Vocabulary()

    # build vocabulary with k-mer algorithm
    for seq in tqdm(data1["seq"].tolist()):
        seq_slide = " ".join(
            [seq[start_idx : start_idx + TOKEN_LEN] for start_idx in range(0, len(seq) - TOKEN_LEN + 1)]
        )
        vocabulary_obj.add_seq(seq_slide)
    vocabulary_obj.save_dict(VOCAB_PATH0)

    le, labels = label_encoder(data["labels"].tolist(), LE_PATH)

    rebuild_data0 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(0, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data0.append(datum)
    save_json(REBUILD_DATA_INDE1_PATH0, rebuild_data0)

    rebuild_data1 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(1, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data1.append(datum)
    save_json(REBUILD_DATA_INDE1_PATH1, rebuild_data1)

    rebuild_data2 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(2, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data2.append(datum)
    save_json(REBUILD_DATA_INDE1_PATH2, rebuild_data2)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
from configs import DATA_PATH, VOCAB_PATH0, VOCAB_PATH1, VOCAB_PATH2, REBUILD_DATA_PATH0, REBUILD_DATA_PATH1, REBUILD_DATA_PATH2, LE_PATH, TOKEN_LEN, PART_NUM, PART_LEN
from utils import save_pkl, save_json, load_json


def label_encoder(y, label_model_path):
    le = LabelEncoder()
    labels = le.fit_transform(y)
    save_pkl(label_model_path, le)
    print("The classes of labels: ", le.classes_)
    return le, labels


class Vocabulary(object):
    def __init__(self):
        self.token2idx = {"<K>": 0}
        self.idx2token = {0: "<K>"}
        self.idx = 1

    def add_token(self, token):
        if token not in self.token2idx:
            self.token2idx[token] = self.idx
            self.idx2token[self.idx] = token
            self.idx += 1

    def add_seq(self, seq):
        for token in seq.split():
            self.add_token(token)

    # transform sequence into number
    def seq2ver(self, seq):
        seq_idxs = [self.token2idx.get(token, self.token2idx["<K>"]) for token in seq.split()]
        return seq_idxs

    def save_dict(self, dict_path):
        save_json(dict_path, self.token2idx)

    def load_dict(self, dict_path):
        self.token2idx = load_json(dict_path)
        self.idx2token = {idx: token for token, idx in self.token2idx.items()}
        self.idx = len(self.token2idx)

    def __call__(self, token):
        return self.token2idx.get(token, self.token2idx["<K>"])

    def __len__(self):
        return self.idx


if __name__ == "__main__":
    data = pd.read_csv("./benchmark.csv")
    data1 = pd.read_csv("./benchmark_all.csv")

    vocabulary_obj = Vocabulary()

    # build vocabulary with k-mer algorithm
    for seq in tqdm(data1["seq"].tolist()):
        seq_slide = " ".join(
            [seq[start_idx : start_idx + TOKEN_LEN] for start_idx in range(0, len(seq) - TOKEN_LEN + 1)]
        )
        vocabulary_obj.add_seq(seq_slide)
    vocabulary_obj.save_dict(VOCAB_PATH0)

    _, labels = label_encoder(data["labels"].tolist(), LE_PATH)

    rebuild_data0 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(0, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data0.append(datum)
    save_json(REBUILD_DATA_PATH0, rebuild_data0)

    rebuild_data1 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(1, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data1.append(datum)
    save_json(REBUILD_DATA_PATH1, rebuild_data1)

    rebuild_data2 = []
    for seq, features, label in tqdm(
        zip(data["seq"].tolist(), data["features"].tolist(), labels.tolist()), total=len(data)
    ):
        seq = (
            seq[: PART_NUM * PART_LEN]
            if len(seq) >= PART_NUM * PART_LEN
            else seq + "#" * (PART_NUM * PART_LEN - len(seq))
        )
        datum = {}
        # transform sequence into number
        for part_start_idx in range(0, PART_NUM):
            part_seq = seq[part_start_idx * PART_LEN : (part_start_idx + 1) * PART_LEN]
            part_seq_slide = " ".join(
                [
                    part_seq[token_start_idx : token_start_idx + TOKEN_LEN]
                    for token_start_idx in range(2, len(part_seq) - TOKEN_LEN + 1, 3)
                ]
            )
            datum[f"#{part_start_idx}"] = vocabulary_obj.seq2ver(part_seq_slide)
        datum["features"] = eval(features)
        datum["label"] = label
        rebuild_data2.append(datum)
    save_json(REBUILD_DATA_PATH2, rebuild_data2)
```
