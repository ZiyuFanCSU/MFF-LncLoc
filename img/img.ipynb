```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 18 
# Data from the image
algorithms = ['our model',
    "Random Forest+3-mer", "Gradient Boosting+3-mer", "Random Forest+4-mer", "Gradient Boosting+4-mer", 
    "Random Forest+5-mer", "Gradient Boosting+5-mer", "Random Forest+6-mer", "Gradient Boosting+6-mer"
]
average_rank1 = [ 0.4375, 0.2889, 0.2874, 0.3059, 0.2990, 0.3101, 0.3342, 0.3400, 0.3204]
average_rank2 = [ 0.4688, 0.3560, 0.3617, 0.3516, 0.3560, 0.3544, 0.3804, 0.3613, 0.3587]

# Assume standard errors (for demonstration purposes)
# These are hypothetical values
std_errors = [0.0125,0.014424, 0.00597537, 0.01313934, 0.01882237, 0.01601768, 0.01661383, 0.0174753, 0.01325639]
std_errors2 = [0.01500, 0.01, 0.015, 0.008, 0.012, 0.009, 0.013, 0.011, 0.014]

# Plotting
fig, ax = plt.subplots(figsize=(10, 7))

# Plot the average rank with error bars
ax.errorbar(average_rank1, algorithms, xerr=[std_errors, std_errors], fmt='o', color='red', ecolor='darkred', capsize=5, markersize=10, elinewidth=2)
ax.errorbar(average_rank2, algorithms, xerr=[std_errors2, std_errors2], fmt='o', color='darkblue', ecolor='blue', capsize=5, markersize=10, elinewidth=2)

# Set labels
ax.set_xlabel('F1-score')


# Set grid
ax.grid(True, linestyle='--', alpha=0.7)

# Set limits
ax.set_xlim(0.2, 0.5)
ax.set_ylim(-1, len(algorithms))

# Show plot
plt.show()
```
